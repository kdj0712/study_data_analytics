## 데이터 선정
- target = age
- features = Term, amount, credit rating

```python
df_RCOCT_extract = df_RCOCT.loc[:, ['age','Term','amount','credit rating']]
df_RCOCT_extract[:2]
```

```python
# 온전한 모델 산출을 위해 결측치 제거
df_RCOCT_extract_droped = df_RCOCT_extract.dropna()
df_RCOCT_extract_droped.isnull().sum()
```

```python
# target과 features를 지정
target = df_RCOCT_extract_droped['age']
features = df_RCOCT_extract_droped[['Term','amount','credit rating']]
target.shape, features.shape
```

```python
from sklearn.model_selection import train_test_split
features_train, features_test, target_train, target_test = train_test_split(features, target, test_size= 0.3, random_state=10)
features_train.shape, features_test.shape, target_train.shape, target_test.shape
```

```python
# 선형회귀 방법 적용
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(features_train, target_train)
```

```python
from sklearn.metrics import r2_score
# train_set : X인 features_train에 대응하는 Y의 값이 출력
target_train_predict = model.predict(features_train)
target_train.shape , target_train_predict.shape # 원래 정답, 학습 후 정답
```

```python
# train_set의 정확도 평가
r2_score(target_train, target_train_predict)
```

```python
# target_set : X인 target_set에 대응하는 Y의 값이 출력
target_test_predict = model.predict(features_test)
target_test.shape, target_test_predict.shape
```

```python
r2_score(target_test, target_test_predict)
```

```python
# pickle 파일 생성
import pickle
with open('RentalCarOfContractType_age_LinearRegression.pkl','wb') as pickle_file :
  pickle.dump(obj=model, file=pickle_file)
```

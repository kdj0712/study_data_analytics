```python
import pandas as pd
```

```python
double_list = [
    [1000, '과자', '2019-12-31', '반품'],
    [2000, '음료', '2020-03-02', '정상'],
    [3000, '아이스크림', '2020-02-03', '정상'],
    [1000, '과자', '2019-12-31', '반품'],
    [4000, '초콜릿', '2020-03-15', '정상'],
    [5000, '사탕', '2020-04-01', '정상'],
    [6000, '빵', '2020-05-20', '반품'],
    [7000, '우유', '2020-06-30', '정상'],
    [8000, '치즈', '2020-07-07', '반품']
]
```

```python
double_columns = ['가격','종류','판매일자','반품여부']
df_sales = pd.DataFrame(data=double_list, columns=double_columns)
# df_sales.info()
df_sales
```

### pd.pivot_table()
- 전체 데이터 = dateFrame
- Index  = group by (unique())
- values = aggregation 대상 컬럼의 이름
- aggfunc = 집계(aggregation)


```python
# 각 row의 개수(record의)
pd.pivot_table(data=df_sales[['가격','종류','반품여부']]
               ,index=['반품여부']
               ,values=['가격']
               ,aggfunc='count')
```

```python
pd.pivot_table(data=df_sales[['가격','종류','반품여부']]
               ,index=['반품여부']
               ,values=['가격']
               ,aggfunc='sum')
```

### pivot_table with Numpy

```python
import numpy as np
pd.pivot_table(data=df_sales[['가격','종류','반품여부']]
               ,index=['반품여부']
               ,values=['가격']
               ,aggfunc=np.sum)
```

```python
df_sales_pivots = pd.pivot_table(data=df_sales[['가격','종류','반품여부']]
               ,index=['반품여부'],values=['가격'],aggfunc=np.sum)
df_sales_pivots
```

```python
#df_sales_plivots = df_sales_pivots.reset_index() 아래와 동일
df_sales_pivots.reset_index(inplace=True)
```

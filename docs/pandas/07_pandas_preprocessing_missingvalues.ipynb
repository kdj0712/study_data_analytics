```python
import pandas as pd
import numpy as np
```

## 전처리(preprocessing)
- 정형 데이터를 기준 : 이상치 / 결측치에 대한 값 처리 방법

```python
# 데이터셋 생성
data = {
    "수치형": [1, None, 3, np.nan, 5],
    "범주형": [None, 'B', 'C', np.nan, 'E'],
    "datetime": pd.date_range(start='2021-01-01', periods=4).insert(3, np.nan)
}
df_temp = pd.DataFrame(data)
df_temp
```

### 결측치(missing values)
- 빠져있는 값 : Null, NaN, None, ""
- 결측치 확인 : pd.isnull().sum()
- 처리 방식
 - 결측치 치환 : 대표값 or 머신러닝 예측값, 열 단위로 적용
 - 결측치 제거 : pd.dropna(subset=[]), 행 단위로 적용

#### 결측치 확인

```python
df_temp.info()
```

```python
df_temp.isnull().sum() # 각 컬럼별 결측치 총 합 표시
```

#### 결측치 처리

##### 1. 행 제거

```python
df_temp.dropna()   # na = Nan
# df_temp.dropna(inplace=True)
df_temp
```

```python
df_temp.dropna(subset=['범주형'])
```

##### 2. 결측치 채우기
- 대표값을 설정해 채용
- 머신러닝(KNN)

```python
df_temp.describe().loc['mean','수치형']
```

```python
df_temp['수치형'].fillna(2.0, inplace=True)
```

```python
df_temp.describe(include=object)
```

```python
df_temp['범주형'].value_counts()
```

```python
df_temp['범주형'].fillna('E', inplace=True)
```

```python
df_temp.describe(include='datetime') # 분포에 많은 영향을 주기 때문에 되도록 삭제
```

```python
df_temp.dropna(inplace=True)
df_temp
```
